# Variable correlation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('world-happiness/2015.csv', sep=',')
df_16 = pd.read_csv('world-happiness/2016.csv', sep=',')
```

```python
df_15.head()
df_15.drop(['Standard Error'], axis=1, inplace=True)
```

```python
df_16.head()
df_16.drop(['Lower Confidence Interval','Upper Confidence Interval'], axis=1, inplace=True)
```

```python
df_15.head()
```

```python
corr_15 = df_15.corr()
corr_16 = df_16.corr()
```

```python
sns.heatmap(corr_16, vmax=1, square=True, annot=True)
plt.show()
```
