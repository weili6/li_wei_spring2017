# Compare on-list country distributions in 2015 and 2016

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('world-happiness/2015.csv', sep=',')
df_16 = pd.read_csv('world-happiness/2016.csv', sep=',')
```

```python
df_15tmp = pd.DataFrame(df_15[['Country','Region','Happiness Score']])
df_15tmp['Year'] = '2015'

df_16tmp = pd.DataFrame(df_16[['Country','Region','Happiness Score']])
df_16tmp['Year'] = '2016'

df_tmp = pd.concat([df_15tmp,df_16tmp])
```

```python
sns.violinplot(x='Region', y='Happiness Score',  data=df_tmp, hue = 'Year', split=True)
plt.xticks(rotation=30)
plt.show()
```

- most regions remain the same
- Central and Eastern Europe clusters between 5-6 in 2016
- Sub-Saharan clusters around 4 in 2016
