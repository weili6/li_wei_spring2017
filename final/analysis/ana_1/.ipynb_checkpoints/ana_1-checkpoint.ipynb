# Analysis 1: Compare regional distributions of countries in World Happiness Reports of 2015 and 2016

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('../../data/2015.csv', sep=',')
df_16 = pd.read_csv('../../data/2016.csv', sep=',')
```

```python
df_15tmp = pd.DataFrame(df_15[['Country','Region','Happiness Score']])
df_15tmp['Year'] = '2015'                                                  #add Year variable

df_16tmp = pd.DataFrame(df_16[['Country','Region','Happiness Score']])
df_16tmp['Year'] = '2016'

df_tmp = pd.concat([df_15tmp,df_16tmp])                                    #concatenate two dataframes
df_tmp.head()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
plt.suptitle('Regional Distribution',fontsize = 10)
sns.violinplot(x='Region', y='Happiness Score',  data=df_tmp, hue = 'Year', split=True, ax = ax)   #violin plot
plt.xticks(rotation=30)
plt.savefig('ana_1_output/plot_1.png', bbox_inches='tight')
plt.show()
```
