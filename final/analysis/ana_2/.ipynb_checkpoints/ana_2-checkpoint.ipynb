# Analysis 2: Rank change

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

```python
df_15 = pd.read_csv('../../data/2015.csv', sep=',')
df_16 = pd.read_csv('../../data/2016.csv', sep=',')
```

```python
def get_region(df):
    if df['Region_2015'].notnull:
        df['Region'] = df['Region_2015']
        return df
    else:
        df['Region'] = df['Region_2016']
        return df
```

```python
df_1516 = pd.merge(df_15, df_16, on='Country', how='inner', suffixes=('_2015', '_2016'))
df_rank = pd.DataFrame(df_1516[['Country','Region_2015','Region_2016','Happiness Rank_2015','Happiness Rank_2016']])
df_rank = get_region(df_rank).drop(['Region_2015','Region_2016'], axis = 1)   #get region for each country
df_rank.head()
```

```python
df_rank['Rank Change'] = df_rank['Happiness Rank_2016'] - df_rank['Happiness Rank_2015']  #calculate the rank change for each country
df_rank = df_rank.sort_values(by = 'Rank Change').reset_index(drop=True)
df_rank.to_csv('ana_2_output/csv_1.csv', sep=',', index=False)
df_rank.head()
```

```python
most_increase = df_rank.head(1)
most_increase.to_csv('ana_2_output/csv_2.csv', sep=',', index=False)
most_increase
```

```python
most_decrease = df_rank.tail(1)
most_decrease.to_csv('ana_2_output/csv_3.csv', sep=',', index=False)
most_decrease
```

```python
df_all = pd.merge(df_15, df_16, on='Country', how='outer', suffixes=('_2015', '_2016'))
df_new = pd.DataFrame(df_all[['Country','Region_2015','Region_2016','Happiness Rank_2015','Happiness Rank_2016']])
df_in = df_new[df_new['Happiness Rank_2015'].isnull()][['Country','Region_2016','Happiness Rank_2016']].reset_index(drop=True)   
df_in.to_csv('ana_2_output/csv_4.csv', sep=',', index=False)
df_in  #new countries in 2016 list
```

```python
df_off = df_new[df_new['Happiness Rank_2016'].isnull()][['Country','Region_2015','Happiness Rank_2015']].reset_index(drop=True)       
df_off.to_csv('ana_2_output/csv_5.csv', sep=',', index=False)
df_off   #countries off 2016 list
```

```python
fig, ax = plt.subplots(figsize=(9,5))
plt.suptitle('Rank Change Distribution',fontsize = 10)
sns.swarmplot(x='Region', y='Rank Change', data=df_rank, ax=ax)
plt.xticks(rotation=30)
plt.savefig('ana_2_output/plot_1.png', bbox_inches='tight')
plt.show()
```
