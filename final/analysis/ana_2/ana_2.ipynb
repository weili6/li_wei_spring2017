# Rank change

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('world-happiness/2015.csv', sep=',')
df_16 = pd.read_csv('world-happiness/2016.csv', sep=',')
```

```python
df_1516 = pd.merge(df_15, df_16, on='Country', how='inner', suffixes=('_2015', '_2016'))
df_rank = pd.DataFrame(df_1516[['Country','Region_2015','Happiness Rank_2015','Happiness Rank_2016']])
```

```python
df_rank['Rank Difference'] = df_rank['Happiness Rank_2016'] - df_rank['Happiness Rank_2015']
df_rank = df_rank.rename(columns={'Region_2015': 'Region'}).sort_values(by = 'Rank Difference').reset_index(drop=True)
df_rank.head()
```

```python
biggest_rise = df_rank.head(1)
biggest_rise
```

```python
biggest_drop = df_rank.tail(1)
biggest_drop
```

```python
sns.swarmplot(x='Region', y='Rank Difference', data=df_rank)
plt.xticks(rotation=30)
plt.show()
```
