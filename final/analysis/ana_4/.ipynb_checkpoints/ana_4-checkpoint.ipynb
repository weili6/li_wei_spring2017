# Analysis 4: Variable correlation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('../../data/2015.csv', sep=',')
df_16 = pd.read_csv('../../data/2016.csv', sep=',')
```

```python
df_15.drop(['Standard Error'], axis=1, inplace=True)
```

```python
df_16.drop(['Lower Confidence Interval','Upper Confidence Interval'], axis=1, inplace=True)
```

```python
corr_15 = df_15.corr()
corr_16 = df_16.corr()
```

```python
fig, ax = plt.subplots(figsize=(9,5))
plt.suptitle('Variable Correlation',fontsize = 10)
sns.heatmap(corr_16, vmax=1, square=True, annot=True, ax=ax)
plt.savefig('ana_4_output/plot_1.png', bbox_inches='tight')
plt.show()
```
