# Analysis 5: Compare mean values of important variables among regions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('../../data/2015.csv', sep=',')
df_16 = pd.read_csv('../../data/2016.csv', sep=',')
```

```python
df_tmp = pd.DataFrame(df_16[['Region','Economy (GDP per Capita)','Family','Health (Life Expectancy)']])
```

```python
fig = plt.figure(figsize=(17, 8))
plt.suptitle('Mean Values of Important Variables Among Regions',fontsize = 10)
ax1 = fig.add_subplot(221)
sns.barplot(x="Region", y="Economy (GDP per Capita)", data=df_tmp, ax=ax1)
plt.xticks(rotation=30)

ax2 = fig.add_subplot(222)
sns.barplot(x="Region", y="Family", data=df_tmp, ax=ax2)
plt.xticks(rotation=30)

ax3 = fig.add_subplot(223)
sns.barplot(x="Region", y="Health (Life Expectancy)", data=df_tmp, ax=ax3)
plt.xticks(rotation=30)
plt.savefig('ana_5_output/plot_1.png', bbox_inches='tight')
plt.show()
```
