```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_15 = pd.read_csv('world-happiness/2015.csv', sep=',')
df_16 = pd.read_csv('world-happiness/2016.csv', sep=',')
```

```python
df_16.head()
```

```python
df_tmp = pd.DataFrame(df_16[['Region','Economy (GDP per Capita)','Family','Health (Life Expectancy)']])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
sns.barplot(x="Region", y="Economy (GDP per Capita)", data=df_tmp, ax=ax1)
plt.xticks(rotation=30)

ax2 = fig.add_subplot(222)
sns.barplot(x="Region", y="Family", data=df_tmp, ax=ax2)
plt.xticks(rotation=30)

ax3 = fig.add_subplot(223)
sns.barplot(x="Region", y="Health (Life Expectancy)", data=df_tmp, ax=ax3)
plt.xticks(rotation=30)
```

```python
plt.show()
```
