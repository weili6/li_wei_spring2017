# Question 3 - Analysis 3
- analyze the trend of the number of comments created over a day
- in order to see when users are most active in a day

```python
import json
import requests
import os
import glob 
import collections
import datetime
import matplotlib.pyplot as plt
```

## Next block is for data download, please skip it

```python
nyt_most_popular_key = os.getenv('auth_key')              # get the dataset of comments by date from NYT API
for i in range(323):
    num = i*25
    url = 'http://api.nytimes.com/svc/community/v3/user-content/by-date.json?date=2017-03-04&offset='+str(num)+'&api-key='+nyt_most_popular_key
    resp = requests.get(url)
    data = json.loads(resp.text)
    with open('../data/comments_by_date/'+str(i+1)+'.json', 'w') as outfile:
        json.dump(data, outfile)
```

```python
filepaths = glob.glob("../data/comments_by_date/*.json", recursive = True)
time_list = []
for name in filepaths:
    with open(name) as file:
        data = json.load(file)
        for i in range(len(data['results']['comments'])):
            time = data['results']['comments'][i]['createDate']
            time_list.append(time)
```

```python
def convert_time(unixtime):                      # define a function to convert unixtime to readable format
    return datetime.datetime.fromtimestamp(int(unixtime)).strftime('%Y-%m-%d %H:%M:%S')
```

```python
new_list = list(map(lambda x: convert_time(x), time_list))         # use lambda function to change the form of time in the list
processed_timelist = list(map(lambda time: time.split(' ',1)[1].split(':',1)[0], new_list))
time_freq = collections.Counter()                                  # use counter to sort the number of comments created by hour
for time in processed_timelist:
    time_freq[time] += 1
od = collections.OrderedDict(sorted(time_freq.items(), key=lambda t: t[0]))
x_list = []
y_list = []
for pair in od.items():
    x_list.append(pair[0][:2]+':00')
    y_list.append(pair[1])
```

```python
plt.figure(figsize=(30,10))                       # plot bar chart
plt.suptitle('The number of new comments vs. the hour of a day', fontsize = 30)
plt.ylabel('number of new comments', fontsize = 20)
plt.xlabel('time', fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 15)
plt.bar(range(len(y_list)), y_list, tick_label=x_list)
plt.savefig('ana_3/ana_3.png')
plt.show() 
```
