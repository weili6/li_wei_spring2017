# Question 2 - Analysis 2
- for a popular article, analyze the trend of new comments created hourly after the article published
- in order to know the lifespan of an article

```python
import json
import requests
import os
import glob
import datetime  
import collections
import matplotlib.pyplot as plt
```

## Next block is for data download, please skip it

```python
nyt_most_popular_key = os.getenv('auth_key')
for i in range(82):
    num = i*25
    url = 'http://api.nytimes.com/svc/community/v3/user-content/url.json?url=http%3A%2F%2Fwww.nytimes.com%2F2017%2F03%2F01%2Fus%2Fpolitics%2Fobama-trump-russia-election-hacking.html&sort=newest&offset='+str(num)+'&api-key='+nyt_most_popular_key
    resp = requests.get(url)
    data = json.loads(resp.text)
    with open('../data/comments_by_url/'+str(i+1)+'.json', 'w') as outfile:
        json.dump(data, outfile)                  # get the dataset of user comments on a specific article from NYT API
```

```python
filepaths = glob.glob("../data/comments_by_url/*.json", recursive = True)
time_list = []
for name in filepaths:
    with open(name) as file:
        data = json.load(file)                    # get the create date of each comments
        for i in range(len(data['results']['comments'])):
            time = data['results']['comments'][i]['createDate']
            time_list.append(time)
            for j in range(len(data['results']['comments'][i]['replies'])):
                time_reply = data['results']['comments'][i]['replies'][j]['createDate']
                time_list.append(time_reply)
```

```python
def convert_time(unixtime):                      # define a function to convert unixtime to readable format
    return datetime.datetime.fromtimestamp(int(unixtime)).strftime('%Y-%m-%d %H:%M:%S')
```

```python
new_list = list(map(lambda x: convert_time(x), time_list))         # use lambda function to change the form of date in the list
processed_timelist = list(map(lambda time: time.split('-',2)[2].split(' ',1)[0]+time.split(' ',1)[1].split(':',1)[0], new_list))
time_freq = collections.Counter()                                  # use counter to sort the number of comments created by hour
for time in processed_timelist:
    time_freq[time] += 1
od = collections.OrderedDict(sorted(time_freq.items(), key=lambda t: t[0]))
x_list = []
y_list = []
for pair in od.items():
    x_list.append('03/'+pair[0][:2]+' '+pair[0][2:4]+':00')
    y_list.append(pair[1])
```

```python
plt.figure(figsize=(80,20))                       # plot bar chart
plt.suptitle('The number of new comments vs. time after the article published', fontsize = 40)
plt.ylabel('number of new comments', fontsize = 20)
plt.xlabel('time', fontsize = 20)
plt.bar(range(len(y_list)), y_list, tick_label=x_list)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.savefig('ana_2/ana_2.png')
plt.show() 
```
